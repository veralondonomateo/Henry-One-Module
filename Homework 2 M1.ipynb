1) Abrir el archivo "Emisiones_CO2.csv" y cargar sus datos en un diccionario.

```python
import os
archivo = open('Emisiones_CO2.csv', 'r', encoding='latin-1')
```

```python
dicc_emisiones = {  'cod_pais' : [],
                    'nom_pais' : [],
                    'region' : [],
                    'anio' : [],
                    'co2' : [],
                    'co2_percapita' : []}
```

```python
x = 0
for i in archivo:
    if x > 0:
        campo = i.split(sep = '|')
        dicc_emisiones["cod_pais"].append(campo[0])
        dicc_emisiones['nom_pais'].append(campo[1])
        dicc_emisiones["region"].append(campo[2])
        dicc_emisiones["anio"].append(campo[3])
        dicc_emisiones["co2"].append(campo[4])
        dicc_emisiones["co2_percapita"].append(campo[5])
    else:
        x += 1
```

```python
archivo.close()
```
